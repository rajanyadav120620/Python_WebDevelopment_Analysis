```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df = pd.read_csv("Desktop/PythonP/IScale/data-export (1).csv")
df = pd.read_csv("data-export (1).csv")
```

```python
# df
df.head()
```

### To remove unname  colum

```python
df.columns = df.iloc[0]
df.head()
```

### I want to drop zero indexing row because it repeat two time

```python
df.columns = df.iloc[0]
df = df.drop(index =0).reset_index(drop =True)
df.columns =["Chanel Group","Data_Hourse","User","Sessions",	"Engaged sessions",	"Average engagement time per session",	"Engaged sessions per user",	"Events per session",	"Engagement rate",	"Event count"]
```

```python
df.head()
```

### To  Check type of data type

```python
df.info()
```

 ### To convert object data to date and time formet

```python
df["Data_Hourse"] = pd.to_datetime(df["Data_Hourse"], format="%Y%m%d%H", errors='coerce')
df.head()
df.info()                                   
```

### To convert object data to nuberical data 

```python
# string is as object data
```

```python
numeric_cols=df.columns.drop(["Chanel Group","Data_Hourse"])
df[numeric_cols] =df[numeric_cols].apply(pd.to_numeric,errors='coerce')
df.head()
```

```python
df.info()
```

### Create a new hour column 

```python
df["Hour"] = df["Data_Hourse"].dt.hour
df.head()
```

```python
df.info()
```

### To find averge,mean,min,max used describe

```python
df.describe()
```

#### Session   and used over time 

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
df.groupby("Data_Hourse")[["Sessions","User"]].sum().plot(ax=plt.gca())
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.groupby("Data_Hourse")[["Sessions", "User"]].sum().plot(ax=plt.gca())
plt.title("Sessions and Users by Data_Hourse")
plt.xlabel("Data_Hourse")
plt.ylabel("Sum")
plt.grid(True)
plt.show()
```

### Total User by Chanel

```python
plt.figure(figsize=(8,5))
# It show some worning
#sns.barplot(data=df, x="Chanel Group", y="User", estimator=np.sum, palette="viridis")
sns.barplot(data=df, x="Chanel Group", y="User", estimator=np.sum, hue="Chanel Group", palette="viridis", legend=False)
plt.title("Totol User by Chanel")
plt.xticks(rotation=45)
plt.show()
```

###  Averge engagement time by chanel

```python
df.head()
```

```python
plt.figure(figsize =(8,5))
sns.barplot(data=df, x="Chanel Group" , y="Average engagement time per session", estimator=np.mean, hue="Chanel Group",  palette="magma")
plt.title("Average engagement time per session")
plt.xticks(rotation=45)
plt.show()
```

### Engagement Rate Disribution ny Chanel

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Chanel Group", y="Engaged sessions per user", hue="Chanel Group", legend=False, palette="coolwarm")
plt.title("Engaged sessions per user")
plt.xticks(rotation=45)
plt.show()
```

# Engeged vs engaged session

```python
df.head()
```

```python
session_df = df.groupby("Chanel Group")[["Sessions","Engaged sessions"]].sum().reset_index()
session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged sessions"]
session_df_melted=session_df.melt(id_vars="Chanel Group", value_vars=["Engaged sessions","Non-Engaged"])
plt.figure(figsize=(8,5))
sns.barplot(data=session_df_melted, x="Chanel Group", y="value", hue="variable")
plt.title("Engaged vs Non_Engaged Session")
plt.xticks(rotation=45)
plt.show()
```

# Traffic by hour and channel

```python
heatmap_data = df.groupby(["Hour", "Chanel Group"])["Sessions"].sum().unstack().fillna(0)
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
plt.title("Traffic by Hour and Channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of Day")

plt.show()
```

# Engagement rate vs session over time

```python
df_plot= df.groupby("Data_Hourse")[["Engagement rate","Sessions"]].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(df_plot["Data_Hourse"], df_plot["Engagement rate"], label="Engagement rate", color="green")
plt.title("Engagement Rate vs Session over Time")
plt.xlabel("Data_Hourse")
plt.legend()
plt.grid(True)
plt.show()
```
